```python
import numpy as np
from sklearn.datasets import make_blobs
```

### Preparing the Dataset

```python
x,y=make_blobs(n_samples=500,centers=3,n_features=2,random_state=3,)
```

```python
import matplotlib.pyplot as plt
```

```python
# x=(x-x.mean())/x.std()
```

```python

plt.grid(True)
plt.scatter(x[:,0],x[:,1])
plt.show()
```

```python
k=3

colors=["green","orange","red"]

clusters={}

for i in range(k):
    center=5*(2*np.random.random((x.shape[1],))-1)
    points=[]
    cluster={
        'center':center,
        'points': points,
        'color':colors[i],
    }
    
    clusters[i]=cluster
```

```python

def distance(x1,x2):
    return np.sqrt(np.sum((x1-x2)**2))
```

```python
def assignPointToClusters(clusters):
    for i in range(x.shape[0]):
        dist=[]
        current_x=x[i]
        for kx in range(k):
            d=distance(current_x,clusters[kx]['center'])
            dist.append(d)

        current_cluster=np.argmin(dist)
        
        clusters[current_cluster]['points'].append(current_x)
```

```python
def plotcluster(clusters):
    for i in range(k):
        
        pts=np.array(clusters[i]['points'])
        
        #plotting of the points
        try:
            plt.grid(True)
            plt.scatter(pts[:,0],pts[:,1],color=clusters[i]['color'])
        except:
            pass
        #plotting of the centers
        
        uk=np.array(clusters[i]['center'])
        plt.scatter(uk[0],uk[1],color="black") 
```

```python
def update(clusters):
    for i in range(k):
        pts=np.array(clusters[i]['points'])
        if pts.shape[0]>0:
            new_mu=pts.mean(axis=0)
            clusters[i]['center']= new_mu
            clusters[i]['points']=[]
            
```

```python
# Initial Run
```

```python
assignPointToClusters(clusters)
plotcluster(clusters)
update(clusters)
```
